```python
import pandas as pd
import datetime
from tqsdk import TqApi, TqAuth
from tqsdk.ta import EMA


api = TqApi(auth=TqAuth("13585930634", "shifangyun199"))

klines = api.get_kline_serial(symbol="DCE.i2305", duration_seconds=60*60, data_length=2000)
klines = klines.drop(klines[klines["id"] < 0].index)
klines["datetime"] = klines["datetime"].apply(lambda x: datetime.datetime.fromtimestamp(x / 1e9))


klines.to_csv("./DCE.i2305_1H.csv", encoding="gbk")

print(klines)
```

```python
import numpy as np
import pandas as pd


np.random.seed(123456)
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()

ts.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randn(1000, 3), index=ts.index, columns=list("ABC"))

df = df.cumsum()

plt.figure()

df.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
